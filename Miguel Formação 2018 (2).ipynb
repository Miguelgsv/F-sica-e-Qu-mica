```python
#Gráficos diretamente
%matplotlib inline
#Melhorar qualidade do gráfico
%config InlineBackend.figure.format = 'svg'

#Para utilizar funções matemáticas
import numpy as np
#Obter gráficos
import matplotlib.pyplot as plt

#gerar valores de x - desde -0 até 10 com 1000 pontos
x = np.linspace(0, 2, 1000)


#Definir o tamanho do gráfico
plt.rcParams['figure.figsize'] = (10,5)

#Definir o tamanho das letras
plt.rcParams.update({'font.size':17})

#Definir o eixo dos y's
plt.yticks([0,1], ['$0$', '$1$'])

#Definir o eixo dos x's
plt.xticks([0, 0.4, 0.8, 1.2, 1.6, 2])

#desenhar a linha horizontal de x
plt.axhline(0, color='black')

#desenhar pontos do gráfico e definir a legenda(label) em notação Latex
plt.plot(x, -0.5*9.8*x*x, label=r'$f(x)=\frac{-1}{2}gt^2$')

#aumentar ao eixo dos y para mostrar valor 1 no eixo
plt.ylim(0, -20)

#Definir o título
plt.title('Queda livre - Posição - (unidades SI)')

#Definir a localização da legenda
plt.legend(loc='best')

#Mostrar o gráfico
plt.show()
```

```python
from sympy import *

# This initialises pretty printing
init_printing()
from IPython.display import display

# This command makes plots appear inside the browser window
%matplotlib inline

t, g = symbols("t g")
y = Function("y(t)")

eqn = Eq(y, (-1/2)*g*t**2)
display(eqn)
```
